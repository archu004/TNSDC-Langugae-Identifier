```python
from langdetect import detect
import matplotlib.pyplot as plt
from langcodes import Language

def identify_language(text):
    try:
        language_code = detect(text)
        language_name = Language.make(language_code).language_name('en')
        return language_name
    except:
        return "Unknown"

def visualize_language_distribution(texts):
    languages = [detect(text) for text in texts]
    language_names = [identify_language(text) for text in texts]
    language_counts = {lang: language_names.count(lang) for lang in set(language_names)}

    plt.figure(figsize=(8, 6))
    plt.bar(language_counts.keys(), language_counts.values())
    plt.xlabel('Languages')
    plt.ylabel('Number of Texts')
    plt.title('Distribution of Languages')
    plt.xticks(rotation=45)
    plt.show()

if __name__ == "__main__":
    # Sample texts for demonstration
    texts = [
        "Bonjour, comment ça va?",
        "Hello, how are you?",
        "Hola, ¿cómo estás?",
        "Ciao, come stai?",
        "Привет, как дела?",
        "こんにちは、お元気ですか？",
        "你好，你好吗？"
    ]

    print("Identifying languages...")
    for text in texts:
        language = identify_language(text)
        print(f"Text: '{text}' -> Language: {language}")

    print("\nVisualizing language distribution...")
    visualize_language_distribution(texts)
```
